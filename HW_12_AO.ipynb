SpaCy


```python
#make a summary from text using nltk library and spacy library
text = 'The history of natural language processing (NLP) generally started in the 1950s, although work can be found from earlier periods. In 1950, Alan Turing published an article titled "Computing Machinery and Intelligence" which proposed what is now called the Turing test as a criterion of intelligence.'
```

```python
#load language model
nlp = spacy.load('en_core_web_sm')
```

```python
# remuve punctuation and make all letters lower case
text = text.lower()
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

#remove punctuation from the string
for element in text:
    if element in punctuations:
        text = text.replace(element, "")
print(f'Text without punctuation: {text}')
```

```python
#prepare text for processing
#tokenize text into sentences
sentences = sent_tokenize(text)
print(f'Sentences: {sentences}')
```

```python
#tokenize text into words
words = word_tokenize(text)
print(f'Words: {words}')
```

```python
#remove stop words
stop_words = stopwords.words('english')
words = [w for w in words if not w in stop_words]
print(f'Words without stop words: {words}')
```

```python
#calculate frequency of words
from collections import Counter
word_freq = Counter(words)
print(word_freq)
```

```python
#calculate frequency of sentences
sent_freq = Counter(sentences)
print(sent_freq)
```

```python
#use spacy to get the most important words and heapq to get the most important sentences
import heapq
from heapq import nlargest
```

```python
#summary of text
summary = ''
summary_sentences = heapq.nlargest(2, sent_freq, key=sent_freq.get)
for sentence in summary_sentences:
    summary += sentence

print(f'Summary: {summary}')
```

```python
#make a summary from text using nltk library
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
import spacy
import heapq
from heapq import nlargest
```

```python
#make a summary from text using nltk library
text_nltk = 'The history of natural language processing (NLP) generally started in the 1950s, although work can be found from earlier periods. In 1950, Alan Turing published an article titled "Computing Machinery and Intelligence" which proposed what is now called the Turing test as a criterion of intelligence.'
```

```python
#load language model nltk
nltk.download('punkt') # one time execution for downloading the package punkt
nltk.download('stopwords') # one time execution for downloading the package stopwords
```

```python
#prepare text for processing by nltk
# remuve punctuation and make all letters lower case
text_nltk = text_nltk.lower()
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

#remove punctuation from the string
for element in text_nltk:
    if element in punctuations:
        text_nltk = text_nltk.replace(element, "")
print(f'Text without punctuation: {text_nltk}')
```

```python
#tokenize text into sentences
sentences_nltk = sent_tokenize(text_nltk)
print(f'Sentences: {sentences_nltk}')
```

```python
#tokenize text into words
words_nltk = word_tokenize(text_nltk)
print(f'Words: {words_nltk}')
```

```python
#remove stop words
stop_words_nltk = set(stopwords.words('english'))
words_nltk = [w for w in words_nltk if not w in stop_words_nltk]
print(f'Words without stop words: {words_nltk}')
```

```python
#calculate frequency of words
from collections import Counter
word_freq_nltk = Counter(words_nltk)
print(word_freq_nltk)
```

```python
#calculate frequency of sentences
sent_freq_nltk = Counter(sentences_nltk)
print(sent_freq_nltk)
```

```python
#summary of text
summary_nltk = ''
summary_sentences_nltk = heapq.nlargest(2, sent_freq_nltk, key=sent_freq_nltk.get)
for sentence in summary_sentences_nltk:
    summary_nltk += sentence
print(f'Summary: {summary_nltk}')
```
